# **Partie 0 : Visualisation d'états, Classe Taquin et Classe Node**

Nous récupérons les classes du TP précédent ainsi que la fonction visualise_state.

```python
from IPython.display import display, HTML

def visualize_state(state):
    """Visualizes the given state of the Taquin using HTML."""
    html = "<table>"
    for row in state:
        html += "<tr>"
        for tile in row:
            if tile == 0:
                html += "<td style='background-color: lightgray; width: 30px; height: 30px; text-align: center; font-size: 20px;'> </td>"  # Blank tile
            else:
                html += f"<td style='background-color: lightblue; width: 30px; height: 30px; text-align: center; font-size: 20px;'>{tile}</td>"
        html += "</tr>"
    html += "</table>"
    display(HTML(html))

class Taquin:
    """
    A class representing the Taquin problem.
    """

    def __init__(self, initial_state, goal_state, size):
        self.initial_state = initial_state
        self.goal_state = goal_state
        self.size = size

    def actions(self, state):
        """Returns the possible actions (moves) from the given state."""
        # Find the position of the blank tile (0)
        row, col = next(
            (r, c)
            for r, row in enumerate(state)
            for c, val in enumerate(row)
            if val == 0
        )

        # Define possible moves (up, down, left, right)
        possible_actions = []
        if row > 0:
            possible_actions.append("up")
        if row < self.size-1:
            possible_actions.append("down")
        if col > 0:
            possible_actions.append("left")
        if col < self.size-1:
            possible_actions.append("right")

        return possible_actions

    def result(self, state, action):
        """Returns the state that results from applying the given action."""
        # Create a copy of the state to avoid modifying the original
        new_state = [list(row) for row in state]

        # Find the position of the blank tile (0)
        row, col = next(
            (r, c)
            for r, row in enumerate(state)
            for c, val in enumerate(row)
            if val == 0
        )

        # Apply the action to move the blank tile
        if action == "up":
            new_state[row][col], new_state[row - 1][col] = (
                new_state[row - 1][col],
                new_state[row][col],
            )
        elif action == "down":
            new_state[row][col], new_state[row + 1][col] = (
                new_state[row + 1][col],
                new_state[row][col],
            )
        elif action == "left":
            new_state[row][col], new_state[row][col - 1] = (
                new_state[row][col - 1],
                new_state[row][col],
            )
        elif action == "right":
            new_state[row][col], new_state[row][col + 1] = (
                new_state[row][col + 1],
                new_state[row][col],
            )

        return new_state

    def is_goal(self, state):
        return state == self.goal_state  # Directly compare with goal_state

    def cost(self, state, action):
        return 1  # Default cost is 1

class Node:
    """
    A node in a search tree.
    __init__: Initializes a node with its state, parent, action, path cost, and depth.
    __repr__: Provides a string representation of the node.
    __lt__: Defines a comparison operator for nodes based on their states.
    expand: Generates child nodes by applying all possible actions.
    child_node: Creates a single child node for a given action.
    solution: Returns the sequence of actions that led to this node.
    path: Returns the path from the root to this node as a list of nodes.
    """

    def __init__(self, state, parent=None, action=None, path_cost = 0):
        self.state = state
        self.parent = parent
        self.action = action
        self.path_cost = path_cost
        self.depth = 0 if parent is None else parent.depth + 1

    # def __repr__(self):
    #     return "<Node {}>".format(self.state)

    # def __lt__(self, other):
    #     return self.state < other.state

    def expand(self, problem):
        """List the nodes reachable in one step from this node."""
        return [
            self.child_node(problem, action)
            for action in problem.actions(self.state)
        ]

    def child_node(self, problem, action):
        """Create a child node by applying the given action."""
        next_state = problem.result(self.state, action)
        next_node = Node(
            next_state,
            parent=self,
            action=action,
            path_cost = self.path_cost + problem.cost(self.state, action),
        )
        return next_node

    def solution(self):
        """Return the sequence of actions to go from the root to this node."""
        return [node.action for node in self.path()[1:]]

    def path(self):
        """Return a list of nodes forming the path from the root to this node."""
        node, path_back = self, []
        while node:
            path_back.append(node)
            node = node.parent
        return list(reversed(path_back))
```

## **1.1 H_1 : Nombre de tuiles mal placées**

### **Exercice 1**

Écrivez une fonction qui calcule le nombre de tuiles mal placées dans un état du jeu de Taquin

```python
def h1(node, problem):
    current_state = node.state
    goal_state = problem.goal_state
    count = 0
    
    for r in range(problem.size):
        for c in range(problem.size):
            tile = current_state[r][c]
            if tile != 0 and tile != goal_state[r][c]:
                count += 1
                
    return count
```

## **1.2 H_2 : Distance de Manhattan**

### **Exercice 3**
Écrivez une fonction qui calcule la distance de Manhattan entre un état et l'état but. La fonction **next()** pourrait être utile pour trouver la position d'une tuile donnée dans l'état but.

```python
def h2(node, problem):
    current_state = node.state
    goal_state = problem.goal_state
    distance = 0
    size = problem.size
    
    # 遍历当前状态的每一个格子
    for r in range(size):
        for c in range(size):
            tile = current_state[r][c]
            
            # 如果不是空块(0)，我们就需要计算它的曼哈顿距离
            if tile != 0:
                # --- 开始查找 tile 在目标状态中的位置 ---
                target_r, target_c = -1, -1
                
                # 遍历目标状态去找这个数字在哪里
                found = False
                for gr in range(size):
                    for gc in range(size):
                        if goal_state[gr][gc] == tile:
                            target_r, target_c = gr, gc
                            found = True
                            break # 找到了就跳出内层循环
                    if found:
                        break # 找到了就跳出外层循环
                # --- 查找结束 ---
                
                # 计算距离并累加
                distance += abs(r - target_r) + abs(c - target_c)
                
    return distance
```

## **1.3 H_3 : Distance de Manhattan avec Conflits Linéaires**
- **$h3$ : Distance de Manhattan avec conflits linéaires.** Cette amélioration de la distance de Manhattan prend en compte les conflits linéaires. Un conflit linéaire se produit lorsque deux conditions se produisent :
  - Deux tuiles sont dans la même ligne ou colonne que leur destination finale.
  - Elles sont dans le mauvais ordre : la tuile de plus petite valeur est à droite ou en dessous de celle ayant une plus grande valeur.

Par exemple, dans la figure 1, l'état initial présente deux conflits linéaires : entre 2 et 1 (ils sont dans la bonne ligne mais 2 est devant 1) et entre 8 et 7. Étant donné un état $E$, $h3$ est défini par :

$$
h3(E) = h2(E) + 2 \times \# \text{Conflits linéaires}
$$

```python
def h3(node, problem):
    # 1. 首先引用 h2 计算基础曼哈顿距离
    # 注意：确保 h2 函数已经在前面定义并运行过
    h2_val = h2(node, problem)

    current_state = node.state
    goal_state = problem.goal_state
    size = problem.size
    linear_conflicts = 0

    # 2. 检测行冲突 (Row Conflicts)
    for r in range(size):
        row_tiles = []
        for c in range(size):
            tile = current_state[r][c]
            # 跳过空块(0)
            if tile != 0:
                # 找到该块在目标状态中的位置
                target_r, target_c = next(
                    (gr, gc)
                    for gr, goal_row in enumerate(goal_state)
                    for gc, goal_val in enumerate(goal_row)
                    if goal_val == tile
                )
                # 如果块现在就在它的目标行中，它可能产生行冲突
                if target_r == r:
                    row_tiles.append((c, target_c))  # 存储 (当前列, 目标列)

        # 检查这一行中的冲突对
        # 冲突定义：两个块都在正确的目标行，但左右顺序反了 (即当前在左边的块，其目标实际上在右边)
        for i in range(len(row_tiles)):
            for j in range(i + 1, len(row_tiles)):
                c_current_i, c_goal_i = row_tiles[i]
                c_current_j, c_goal_j = row_tiles[j]

                # 因为我们是按当前列顺序添加的，所以 tile_i 在 tile_j 左边
                # 如果 tile_i 的目标列竟然比 tile_j 的目标列还大（在右边），说明发生冲突
                if c_goal_i > c_goal_j:
                    linear_conflicts += 1

    # 3. 检测列冲突 (Column Conflicts)
    for c in range(size):
        col_tiles = []
        for r in range(size):
            tile = current_state[r][c]
            if tile != 0:
                target_r, target_c = next(
                    (gr, gc)
                    for gr, goal_row in enumerate(goal_state)
                    for gc, goal_val in enumerate(goal_row)
                    if goal_val == tile
                )
                # 如果块现在就在它的目标列中，它可能产生列冲突
                if target_c == c:
                    col_tiles.append((r, target_r))  # 存储 (当前行, 目标行)

        # 检查这一列中的冲突对
        for i in range(len(col_tiles)):
            for j in range(i + 1, len(col_tiles)):
                r_current_i, r_goal_i = col_tiles[i]
                r_current_j, r_goal_j = col_tiles[j]

                # 如果上面的块(i) 目标其实在下面(比j大)，说明发生冲突
                if r_goal_i > r_goal_j:
                    linear_conflicts += 1

    # h3 = h2 + 2 * 线性冲突数
    return h2_val + 2 * linear_conflicts
```

## **1.4 H_4 : Nombre de Tuiles Mal Placées Pondérée**
- **$h4$ : Nombre de tuiles mal placées pondérée.** Elle compte le nombre de tuiles mal placées par rapport à l'état final, tout en tenant compte du fait que certaines tuiles sont plus difficiles à déplacer que d'autres. On considère :
    - La tuile centrale (5) a un poids de $0.8$.
    - Les tuiles des coins (1, 3, 7) ont un poids de $1.2$.
    - Les tuiles des bords (2, 4, 6, 8) ont un poids de $1$.
    
Pour l'état initial de la figure 1, nous obtenons :

$$
h4(E) = 1 + 1.2 + 1 + 0.8 + 1 + 1.2 + 1.2 = 7.4
$$

```python
def h4(node, problem):
    current_state = node.state
    goal_state = problem.goal_state
    weighted_misplaced = 0
    size = problem.size

    # 定义每个数字对应的权重
    # 中心块 (5): 0.8
    # 角落块 (1, 3, 7): 1.2  (注意题目里没提9，因为是3x3拼图只有0-8，不过通常角落还有个位置)
    # 边缘块 (2, 4, 6, 8): 1.0
    # 我们用一个字典来存这些权重，方便查找
    weights = {
        5: 0.8,
        1: 1.2,
        3: 1.2,
        7: 1.2,  # 题目只列了1,3,7，如果还有其他角落块可视情况添加
        2: 1.0,
        4: 1.0,
        6: 1.0,
        8: 1.0,
    }

    for r in range(size):
        for c in range(size):
            tile = current_state[r][c]

            # 如果不是空块，且位置不对（即错位）
            if tile != 0 and tile != goal_state[r][c]:
                # 获取该数字对应的权重，如果字典里没有（比如大于8的情况），默认给1.0
                weight = weights.get(tile, 1.0)
                weighted_misplaced += weight

    return weighted_misplaced
```

# **Partie 2 : Algorithme A étoile**

### **Exercice 5**

Créer une fonction A_etoile qui prend un objet problem en entrée ainsi qu'une heuristique et exécute l'algorithme A.*

Votre fonction doit :

1. Retourner le nœud objectif trouvé ainsi que le nombre de nœuds explorés.
2. Retourner None si l'algorithme explore tout l'arbre sans trouver le nœud objectif.
3. Prendre en compte un budget d'exploration pour arrêter l'algorithme si aucune solution n'est trouvée après plusieurs itérations (utile pour les grands problèmes). Le budget sera également un paramètre d'entrée de la fonction, avec float('inf') comme valeur par défaut.

Pour la **frontière** et l'ensemble des **nœuds déjà explorés**, vous pouvez utiliser des listes. Cependant, la frontière doit contenir les nœuds à explorer ainsi que leurs coûts (coût réel + heuristique). Pour trouver l'élément (nœud, coût) de coût minimal dans la frontière, effectuez une boucle **for** pour trouver l'indice, puis utilisez pop.

Pour ajouter un élément à une liste, vous pouvez utiliser **liste.append(element)**.

Rappelez-vous que la vérification **is_goal** doit être effectuée avant d'explorer les enfants.

```python
def A_etoile(problem, heuristic, budget=float('inf')):

    start_node = Node(problem.initial_state)
    
    start_f = start_node.path_cost + heuristic(start_node, problem)
    
    frontier = [(start_node, start_f)]
    
    explored = []
    
    nodes_explored_count = 0
    
    while frontier:
        if nodes_explored_count >= budget:
            return None, nodes_explored_count
        
        min_index = 0
        for i in range(1, len(frontier)):
            if frontier[i][1] < frontier[min_index][1]:
                min_index = i
        
        current_node, current_f = frontier.pop(min_index)
        
        if problem.is_goal(current_node.state):
            return current_node, nodes_explored_count #Return the node and the number of nodes explored
        
        explored.append(current_node.state)
        nodes_explored_count += 1
        
        for child in current_node.expand(problem):
            child_state = child.state
            
            if child_state in explored:
                continue
            
            child_f = heuristic(child, problem)
            
            in_frontier_index = -1
            for i in range(len(frontier)):
                if frontier[i][0].state == child_state:
                    in_frontier_index = i
                    break
            
            if in_frontier_index != -1:
                if child_f < frontier[in_frontier_index][1]:
                    frontier.pop(in_frontier_index)
                    frontier.append((child, child_f))
            else:
                frontier.append((child, child_f))
                
    return None, nodes_explored_count
```
