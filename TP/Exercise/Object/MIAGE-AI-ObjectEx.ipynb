## A. Génération du réseau

```python
from typing import Type


class ReseauSocial:
    # 1. Définir la classe réseau social et son constructeur 
    def __init__(self):
        self.__reseau = {}
    
    # 2. Ajouter une méthode adduser(self,user)
    def adduser(self,user):
        if not isinstance(user, str):
            raise TypeError("user must be a string")
        if user not in self.__reseau:
            self.__reseau[user] = []

    # 3. Redéfinissez la méthode prédéfinie 
    def __str__(self):
        return str(self.__reseau)

    # 5. Ajouter une méthode addlink(self, user1, user2)
    def addlink(self, user1, user2):
        if (not isinstance(user1, str)) or (not isinstance(user2, str)):
            raise TypeError("User must be a string")
        if (user1 not in self.__reseau) or (user2 not in self.__reseau):
            raise ValueError("User not in the social web")
        if user1 == user2:
            raise ValueError("Cannot add friend to self")
        if (user2 in self.__reseau[user1]) or (user1 in self.__reseau[user2]):
            raise ValueError("User is already in the friend list")
        self.__reseau[user1].append(user2)
        self.__reseau[user2].append(user1)

    # 6. Revoie tous les utilisateurs
    def getusers(self):
        return self.__reseau.keys()
    
```

```python
# 4. Test de la classe ReseauSocial
rs = ReseauSocial()
rs.adduser("Adam")
rs.adduser("Bob")
rs.adduser("Cris")
print(rs)
```
