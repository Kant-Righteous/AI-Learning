#### 15. [Question ouverte] Utiliser la librairie matplolib pour des graphiques intéressants et informatifs.

```python
import matplotlib.pyplot as plt

pk['Type 1'].isna().sum()
```

```python
type1Gr = pk.groupby('Type 1')
print(len(type1Gr))
```

```python
type1Gr['Total'].mean().head(5).plot.bar()
```

```python
import matplotlib.pyplot as plt

# figsize: 图片大小 (宽, 高)
# color: 柱子颜色 (可以使用十六进制或颜色名，如 'skyblue', '#FF9999')
# edgecolor: 柱子边框颜色
# rot: x轴标签旋转角度
# width: 柱子宽度
type1Gr['Total'].mean().head(5).plot.bar(
    figsize=(10, 6),
    color='skyblue',
    edgecolor='black',
    rot=0,
    width=0.6
)

# 添加标题和坐标轴标签
plt.title('Average Total Stats by Pokemon Type 1 (Top5)', fontsize=16, fontweight='bold')
plt.xlabel('Type 1', fontsize=12)
plt.ylabel('Average Total Stats', fontsize=12)

# 添加水平网格线（axis='y'），并设置透明度
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
